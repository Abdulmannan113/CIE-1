```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_csv('/content/mtcars.csv')
df
```

```python
# Histogram of 'mpg' (Miles per gallon)
plt.figure(figsize=(8, 5))
plt.hist(df['mpg'], bins=10, edgecolor='black', alpha=0.7)
plt.xlabel('Miles per gallon (mpg)')
plt.ylabel('Frequency')
plt.title('Histogram of Miles per gallon (mpg)')
plt.grid(True)
plt.show()
```

```python
# Scatter plot of 'weight' vs. 'mpg'
plt.figure(figsize=(8, 5))
sns.scatterplot(x='wt', y='mpg', data=df, alpha=0.7)
plt.xlabel('Weight of the car')
plt.ylabel('Miles per gallon (mpg)')
plt.title('Scatter Plot: Weight vs. Miles per gallon (mpg)')
plt.grid(True)
plt.show()
```

```python

import pandas as pd
import matplotlib.pyplot as plt
pd.read_csv('/content/mtcars.csv')
transmission_counts = df['am'].value_counts()
transmission_counts.plot(kind='bar')
plt.title('Frequency Distribution of Transmission Type')
plt.xlabel('Transmission Type')
plt.ylabel('Frequency')
plt.show()
```

```python
# Box plot of 'mpg'
plt.figure(figsize=(8, 5))
sns.boxplot(x=df['mpg'])
plt.xlabel('Miles per gallon (mpg)')
plt.title('Box Plot of Miles per gallon (mpg)')
plt.grid(True)
plt.show()
```
